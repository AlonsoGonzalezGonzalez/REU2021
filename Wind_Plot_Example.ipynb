## Example Notebook for using a polar wind plotting function for circular polar plot"

```python
import matplotlib.pyplot as plt     # For plotting
import xarray as xr                 # For dealing with netCDF data
from polar_stereo_function import circular_polar_wind_plot
```

```python
file   = 'WACCMX+DART_UVTGPH_2009011900-2009030523.nc'
ds=xr.open_dataset(file)

#Making U and V data arrays for the first height and the 10th time step (index 9)
#The original data used a count on the dimensions not the values of the dimesions, these data arrays fix that.
data_u = xr.DataArray(ds.U[0,9,:,:], coords=[ds.LATITUDE, ds.LONGITUDE], dims=["lat", "lon"])
data_v = xr.DataArray(ds.V[0,9,:,:], coords=[ds.LATITUDE, ds.LONGITUDE], dims=["lat", "lon"])
```

## Using the stream plot version

```python
#Example of plotting using stream plot
circular_polar_wind_plot(data_u, data_v, 65,'Surface Winds', plot_type='stream')
```

```python
#Change the look of the stream plot by changing the density of lines, default is 1
circular_polar_wind_plot(data_u, data_v, 65, 'Surface Winds', plot_type='stream', s_density=4)
```

## Using the wind barb version

```python
#Example of plotting using wind barb plot
circular_polar_wind_plot(data_u, data_v, 65, 'Surface Winds', plot_type='barb')
```

```python
#Example of plotting using wind barb plot, change the look using the linewidth of the barbs, default is 0.95
circular_polar_wind_plot(data_u, data_v, 65, 'Surface Winds', plot_type='barb', b_linewidth=0.1)
```

## Using the Quiver version of the plot

```python
#Example of plotting using quiver plot
circular_polar_wind_plot(data_u, data_v, 65, 'Surface Winds', plot_type='quiver')
```

```python
#Example of plotting using quiver plot, change the look of the plot with the scale parameter, default is 0.5
circular_polar_wind_plot(data_u, data_v, 65, 'Surface Winds', plot_type='quiver', q_scale = 2000)
```
