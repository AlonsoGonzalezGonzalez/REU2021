```python
import matplotlib.pyplot as plt     # For plotting
import xarray as xr                 # For dealing with netCDF data
from polar_stereo_function import circular_polar_plot
import numpy as np
```

```python
file   = 'WACCMX+DART_UVTGPH_2009011900-2009030523.nc'
ds=xr.open_dataset(file)

#Making U and V data arrays for the first height and the 10th time step (index 9)
#The original data used a count on the dimensions not the values of the dimesions, these data arrays fix that.
data_u = xr.DataArray(ds.U[0,9,:,:], coords=[ds.LATITUDE, ds.LONGITUDE], dims=["lat", "lon"])
data_v = xr.DataArray(ds.V[0,9,:,:], coords=[ds.LATITUDE, ds.LONGITUDE], dims=["lat", "lon"])

circular_polar_plot(data_u, 50, 'jet', 'Zonal Winds',
                    cbar_label='m/s', levels_=np.arange(-40,100,1), cbar_levels_=np.arange(-20, 100, 25),
                   add_text=True, text_lat=-150, text_lon=58, text_string="K", text_size=24 )
```
